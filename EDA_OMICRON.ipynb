```python
# See Content Of The Dataset
dataset = '/content/drive/MyDrive/EXDATA/omicron.csv'
data = pd.read_csv(dataset)
data
```

```python
# Information About All Values Of Dataset
data.info()
```

```python
# Check Mean, Min, Max, Std, Count, Etc. Of Numeric Values Of Dataset
data.describe()
```

```python
# Change Date, Object To Datetime
data['Date'] = pd.to_datetime(data['Date'])
```

```python
# visualisasi data kasus omicron berdasarkan tanggal dengan bar plot
plt.figure(figsize = (20,8))
plt.bar(data['Date'], data['Percentage of Omicron'], color = 'blue', width = 5)
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16)
plt.title("Omicron Case Distributaion (Date By Date)", fontsize = 25)
plt.xlabel("Dates", fontsize = 20)
plt.ylabel("Omicron Case", fontsize = 20)
plt.show()
```

```python
# menghapus data omicron yang presentase nya kurangd ari 0.99% dan mengurutkan datadari yang terbesar
ndf = data[data['Percentage of Omicron'] > 0.99].sort_values(by = 'Percentage of Omicron', ascending = False)
ndf
```

```python
# visualisasi data distribusi omicron yang di urutkan berdasarkan count of country
plt.figure(figsize = (20,8))
sns.histplot(ndf['Percentage of Omicron'], color = 'blue')
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16)
plt.title("Omicron Case Visualization", fontsize = 25)
plt.xlabel("Percentage Of Omicron", fontsize = 20)
plt.ylabel("Count Of Country", fontsize = 20)
plt.show()
```

```python
# Visualize 10 Country That Have Maximum Omicron Casees
plt.figure(figsize = (20,8))
plt.bar(ndf['Country'][:10], ndf['Percentage of Omicron'][:10], color = 'blue')
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16)
plt.title("10 Country With The Highest No Of Omicron Case", fontsize = 25)
plt.xlabel("Country", fontsize = 20)
plt.ylabel("Omicron Case Percentage", fontsize = 20)
plt.show()
```
